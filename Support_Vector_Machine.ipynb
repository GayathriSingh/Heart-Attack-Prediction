```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_data=pd.read_csv('heart_attack_prediction_dataset.csv')
```

```python
heart_data.head()
```

```python
heart_data.info()
```

```python
heart_data.isnull().sum()
```

```python
heart_data.drop(['Income','Country','Continent','Hemisphere'],axis=1,inplace=True)
```

```python
heart_data.head()
```

```python
heart_data.info()
```

```python
heart_data.drop(['Patient ID','Sedentary Hours Per Day','Physical Activity Days Per Week'],axis=1,inplace=True)
```

```python
heart_data.info()
```

```python
heart_data.head()
```

```python
Gender=pd.get_dummies(heart_data['Sex'],drop_first=True)
Gender.head()
```

```python
diet=pd.get_dummies(heart_data['Diet'],drop_first=True)
diet.head()
```

```python
heart_data=pd.concat([heart_data,Gender,diet],axis=1)
```

```python
heart_data.head()
```

```python
heart_data.drop(['Sex','Diet'],axis=1,inplace=True)
```

```python
heart_data.info()
```

```python
blood_pressure_split = heart_data['Blood Pressure'].str.split('/', expand=True)
heart_data['Systolic'] = blood_pressure_split[0].astype(int)
heart_data['Diastolic'] = blood_pressure_split[1].astype(int)
```

```python
heart_data.drop(['Blood Pressure'],axis=1,inplace=True)
```

```python
heart_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = heart_data.drop('Heart Attack Risk', axis=1)
y =heart_data['Heart Attack Risk']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
```
